Creating Embeddings

```python
import json
import numpy as np
from sentence_transformers import SentenceTransformer

# Load dataset
with open("interview_dataset.json", "r") as f:
    interview_data = json.load(f)

# Load embedding model (lightweight + strong)
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# Extract only questions for embedding
questions = [item["question"] for item in interview_data]

# Generate embeddings
question_embeddings = embedding_model.encode(questions)

print("Embedding shape:", question_embeddings.shape)
```

Creating FAISS Index

```python
import faiss

# Get embedding dimension
dimension = question_embeddings.shape[1]

# Create FAISS index (L2 distance)
index = faiss.IndexFlatL2(dimension)

# Convert embeddings to float32 (FAISS requirement)
question_embeddings = np.array(question_embeddings).astype("float32")

# Add embeddings to index
index.add(question_embeddings)

print("Number of vectors in index:", index.ntotal)
```

Create Semantic Search Functions

```python
def retrieve_questions(query, top_k=3):
    # Convert query to embedding
    query_vector = embedding_model.encode([query]).astype("float32")

    # Search in FAISS
    distances, indices = index.search(query_vector, top_k)

    results = []
    for idx in indices[0]:
        results.append(interview_data[idx])

    return results
```

```python
results = retrieve_questions("Explain bias variance tradeoff", top_k=2)

for r in results:
    print("Skill:", r["skill"])
    print("Question:", r["question"])
    print("-" * 50)
```

Loading Lightweight LLM

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")
llm_model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")

def llm_generate_text(prompt, max_new_tokens=256):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = llm_model.generate(**inputs, max_new_tokens=max_new_tokens)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

print("LLM Loaded Successfully (manual setup)")
```

RAG Prompt function

```python
def generate_interview_question(skill_query):

    # Step 1: Retrieve relevant questions
    retrieved = retrieve_questions(skill_query, top_k=2)

    context = ""
    for item in retrieved:
        context += f"""
        Question: {item['question']}
        Expected Points: {', '.join(item['expected_points'])}
        """

    # Step 2: Create RAG Prompt
    prompt = f"""
    You are an AI interview copilot.
    Using the context below, generate ONE structured interview question
    along with expected key answer points.

    Context:
    {context}

    Output format:
    Question:
    Expected Points:
    Follow-up:
    """

    response = llm(prompt)[0]["generated_text"]
    return response
```

```python
import re

def evaluate_answer(question, user_answer):

    prompt = f"""
    You are an AI interview evaluator.

    Question:
    {question}

    Candidate Answer:
    {user_answer}

    Evaluate strictly in this format:

    Technical Depth: <number between 1-5>
    Clarity: <number between 1-5>
    Confidence: <number between 1-5>
    Feedback: <short improvement suggestion>
    """

    response = llm(prompt)[0]["generated_text"]

    # Extract scores using regex
    tech = re.search(r"Technical Depth:\s*(\d)", response)
    clarity = re.search(r"Clarity:\s*(\d)", response)
    confidence = re.search(r"Confidence:\s*(\d)", response)

    tech_score = int(tech.group(1)) if tech else 0
    clarity_score = int(clarity.group(1)) if clarity else 0
    confidence_score = int(confidence.group(1)) if confidence else 0

    return {
        "technical_depth": tech_score,
        "clarity": clarity_score,
        "confidence": confidence_score,
        "raw_feedback": response
    }
```

```python
def run_interview(skill, num_questions=3):

    total_tech = 0
    total_clarity = 0
    total_confidence = 0

    print(f"\nStarting AI Interview for: {skill}\n")

    for i in range(num_questions):
        print(f"\n--- Question {i+1} ---")

        generated = generate_interview_question(skill)
        print("\nInterview Question:")
        print(generated)

        user_answer = input("\nYour Answer: ")

        evaluation = evaluate_answer(generated, user_answer)

        print("\nEvaluation:")
        print(evaluation["raw_feedback"])

        total_tech += evaluation["technical_depth"]
        total_clarity += evaluation["clarity"]
        total_confidence += evaluation["confidence"]

    print("\n===== FINAL REPORT =====")

    avg_tech = total_tech / num_questions
    avg_clarity = total_clarity / num_questions
    avg_confidence = total_confidence / num_questions

    print(f"Average Technical Depth: {avg_tech:.2f}")
    print(f"Average Clarity: {avg_clarity:.2f}")
    print(f"Average Confidence: {avg_confidence:.2f}")
```

```python
run_interview("Machine Learning")
```
